```python
import cv2
import numpy as np
from skimage.feature import hog
from skimage import color
import matplotlib.pyplot as plt

# Load image
image = cv2.imread("PRO.jpg")  # Load original image
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB for proper display

# Convert to grayscale
gray = color.rgb2gray(image)  

# Compute HOG features
features, hog_image = hog(gray, pixels_per_cell=(8, 8), cells_per_block=(2, 2), 
                          orientations=9, visualize=True, block_norm="L2-Hys")

# Plot original, grayscale, and HOG images
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(gray, cmap="gray")
ax[1].set_title("Grayscale Image")
ax[1].axis("off")

ax[2].imshow(hog_image, cmap="gray")
ax[2].set_title("HOG Features")
ax[2].axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import color, exposure

# Load image
image = cv2.imread("IM.jpg")  
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for proper display

# Convert to grayscale
gray = color.rgb2gray(image)

# Compute HOG features
features, hog_image = hog(gray, pixels_per_cell=(8, 8), cells_per_block=(2, 2),
                          orientations=9, visualize=True, block_norm="L2-Hys")

#  Option 1: Increase contrast with histogram equalization
hog_image_eq = exposure.equalize_hist(hog_image)

#  Option 2: Fine-tune contrast with rescaling (try adjusting `in_range`)
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 0.1))

# Create a figure with 4 subplots
fig, ax = plt.subplots(1, 4, figsize=(20, 5))

# Original Image
ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

# Grayscale Image
ax[1].imshow(gray, cmap="gray")
ax[1].set_title("Grayscale Image")
ax[1].axis("off")

# HOG Features (Histogram Equalized)
ax[2].imshow(hog_image_eq, cmap="gray")
ax[2].set_title("HOG Features (Equalized)")
ax[2].axis("off")

# HOG Features (Rescaled Intensity)
ax[3].imshow(hog_image_rescaled, cmap="gray")
ax[3].set_title("HOG Features (Rescaled)")
ax[3].axis("off")

plt.show()
```
